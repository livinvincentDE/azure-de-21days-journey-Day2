Writing the logs into a file

```python
from pyspark.sql import Row
import json

payload_str = payload if 'payload' in locals() else "{}"
data = json.loads(payload_str)

df = spark.createDataFrame([Row(
    file_name = data.get("file_name", "unknown_file"),
    status = data.get("status", "UNKNOWN"),
    time = data.get("time", "unknown_time"),
    run_id = data.get("run_id", "unknown_run"),
    pipeline_name = data.get("pipeline_name", "unknown_pipeline"),
    row_count = int(data.get("row_count", 0))
)])

# Enable schema evolution
spark.conf.set("spark.databricks.delta.schema.autoMerge.enabled", "true")

df.write \
  .mode("append") \
  .option("mergeSchema", "true") \
  .format("delta") \
  .saveAsTable("pipeline_logs")

df.show()
```
